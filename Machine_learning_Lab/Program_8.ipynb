```python
import pandas as pd
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
```

# modeling 

```python
model = BayesianNetwork([
    ('age','heartdisease'),
     ('sex','heartdisease'),
     ('cp','heartdisease'),
      ('exang','heartdisease'),
    ('heartdisease','restecg'),
    ('heartdisease','chol'),
    
    
])
```

```python
model.fit(data,estimator=MaximumLikelihoodEstimator) #MaximumLikelihoodEstimator uses  statistics to find the max parameter
#                                                      of observed data
```

# inferencing from model

```python
infer =VariableElimination(model)
```

```python
print(infer.query(variables=['heartdisease'],evidence={'restecg':1}))
```

```python
print(infer.query(variables=['heartdisease'],evidence={'cp':2}))
```
